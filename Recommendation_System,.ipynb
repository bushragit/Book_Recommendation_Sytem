- the code is counting the number of occurrences of each title in the "Book-Title" column of the book_ratings DataFrame, and returning the count of ratings for each title in the "Book-Rating" column

```python
num_of_ratings = book_ratings.groupby(["Book-Title"]).count()["Book-Rating"].reset_index()
num_of_ratings.rename(columns={"Book-Rating":"num-ratings"}, inplace=True)
```

.com- the code is counting the number of occurrences of each title in the "Book-Title" column of the book_ratings DataFrame, and returning the mean of ratings for each title in the "Book-Rating" column

```python
avg_ratings = book_ratings.groupby(["Book-Title"]).mean()["Book-Rating"].reset_index()
avg_ratings.rename(columns={"Book-Rating":"avg-ratings"}, inplace=True)
```

```python
avg_ratings.head()
```

- merge num_of_ratings columns and avg_ratings on Book_Title

```python
popular_df = num_of_ratings.merge(avg_ratings,on='Book-Title')
```

# collabrative filtering based recommender system

- take users which provide ratings on >= 2oo books

```python
x= book_ratings.groupby("User-ID").count()["Book-Rating"]>200
```

```python
book_ratings.shape
```

```python
x.shape
```

```python
literate = x[x].index
```

```python
filtered_ratings = book_ratings[book_ratings["User-ID"].isin(literate)]
```

  - take books which have atleast 50 or more ratings

```python
y= book_ratings.groupby("Book-Title").count()["Book-Rating"]>50
```

```python
famous_books= y[y].index
```

```python
final = filtered_ratings[filtered_ratings["Book-Title"].isin(famous_books)]
```

```python
final.drop_duplicates()
```

### now make pivot table

```python
pt= final.pivot_table(index= "Book-Title" ,columns="User-ID", values="Book-Rating")
```

```python
pt
```

```python
pt.fillna(0, inplace=True)
```

```python
pt.tail(90)
```

```python
- calculating similarity between users is to compute the cosine similarity between their rating vectors. 
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score= cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
define 
```

-  Python function that recommends similar books based on a book name provided as an input parameter.

```python
def recommend(book_name):
    index= np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1], reverse=True)[1:6]
    data = []
    for i in similar_items:
        items = []
        temp_df = books[books["Book-Title"]==pt.index[i[0]]]
        items.extend(list(temp_df.drop_duplicates("Book-Title")["Book-Title"].values))
        items.extend(list(temp_df.drop_duplicates("Book-Title")["Book-Author"].values))
        items.extend(list(temp_df.drop_duplicates("Book-Title")["Image-URL-M"].values))

        data.append(items)

    return data
    
 
```

```python
- Now see the results,
```
